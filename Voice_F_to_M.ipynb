```python
from gtts import gTTS
from IPython.display import Audio
```

```python
text_to_speech = gTTS('''The IAF's fleet of jets is a mix of multirole fighters, attack aircraft, and aircraft for surveillance and transport,
                          each playing a critical role in Indias national defense strategy. Among the most renowned of these are the Sukhoi 
                          Su-30MKI, HAL Tejas, Dassault Rafale, and Mig-29, all of which contribute to the IAF's ability to respond swiftly 
                          and decisively to both regional and global threats.''',lang='hi',tld='com')

text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay= False)
```

```python
import pyttsx3
from IPython.display import Audio

text = '''The IAF's fleet of jets is a mix of multirole fighters, attack aircraft, and aircraft for surveillance and transport,
                          each playing a critical role in Indias national defense strategy. Among the most renowned of these are the Sukhoi 
                          Su-30MKI, HAL Tejas, Dassault Rafale, and Mig-29, all of which contribute to the IAF's ability to respond swiftly 
                          and decisively to both regional and global threats.'''
audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

# change the voices
voice = audio.getProperty('voices')

# 0 for male and 1 for female
audio.setProperty('voice', voice[0].id)      # for male voice   
#audio.setProperty('voice', voice[1].id)       # for female voice

# text-to speech conversion
audio.say(text)

# save the audio file
audio.save_to_file(text, 'test_male_Voice.mp3')
#audio.save_to_file(text, 'test_female_Voice.mp3')

audio.runAndWait()
```
